```python
# Gerekli kütüphaneleri içe aktarma
import yfinance as yf
import pandas as pd

# 1. Hisseleri tanımla
tickers = ['A1CAP.IS', 'ADESE.IS','ADGYO.IS', 'AEFES.IS', 'AFYON.IS','AGHOL.IS', 'AGESA.IS', 'AGROT.IS', 'AGYO.IS','AHGAZ.IS', 'AHSGY.IS', 'AKBNK.IS', 'AKCNS.IS', 'AKENR.IS', 'AKFGY.IS', 'AKFYE.IS', 'AKGRT.IS', 'AKGRT.IS', 'AKMGY.IS', 'AKMGY.IS', 'AKSEN.IS', 'AKSGY.IS','AKSUE.IS', 'AKYHO.IS', 'ALARK.IS','ALBRK.IS', 'ALCAR.IS', 'ALCTL.IS', 'ALFAS.IS', 'ALKIM.IS', 'ALKLC.IS', 'ALKA.IS', 'ALMAD.IS', 'ALVES.IS', 'ALTNY.IS', 'ANGEN.IS', 'ANELE.IS', 'ANHYT.IS',"ANSGR.IS", 'ARCLK.IS', 'ARDYZ.IS', 'ARENA.IS', 'ARSAN.IS', 'ARTMS.IS', 'ARZUM.IS', 'ASELS.IS', 'ASGYO.IS', 'ASTOR.IS', 'ATAGY.IS', 'ATAKP.IS', 'ATATP.IS', 'ATEKS.IS', 'ATLAS.IS', 'ATSYH.IS', 'AVHOL.IS', 'AVGYO.IS', 'AVOD.IS', 'AVPGY.IS', 'AVTUR.IS', 'AYDEM.IS', 'AYEN.IS', 'AYES.IS', 'AYGAZ.IS', 'AZTEK.IS', 'BAGFS.IS', 'BAHKM.IS', 'BAKAB.IS', 'BALAT.IS', 'BANVT.IS', 'BARMA.IS', 'BASCM.IS', 'BASGZ.IS', 'BAYRK.IS', 'BEGYO.IS', 'BEYAZ.IS', 'BFREN.IS', 'BIMAS.IS','BINHO.IS', 'BIOEN.IS', 'BIENY.IS', 'BIGCH.IS', 'BIZIM.IS', 'BJKAS.IS','BLCYT.IS', 'BMSCH.IS', 'BMSTL.IS', 'BNTAS.IS', 'BOBET.IS', 'BOSSA.IS', 'BORLS.IS', 'BORSK.IS', 'BRISA.IS', 'BRKSN.IS', 'BRKVY.IS', 'BRLSM.IS', 'BRMEN.IS', 'BRYAT.IS', 'BSOKE.IS', 'BTCIM.IS', 'BURCE.IS', 'BURVA.IS', 'BUCIM.IS', 'BVSAN.IS', 'BYDNR.IS','CIMSA.IS', 'CANTE.IS', 'CASA.IS', 'CATES.IS', 'CCOLA.IS', 'CEOEM.IS', 'CELHA.IS', 'CEMAS.IS', 'CEMTS.IS', 'CEMZY.IS', 'CLEBI.IS',   'CONSE.IS', 'COSMO.IS', 'CRDFA.IS', 'CRFSA.IS', 'CVKMD.IS', 'CWENE.IS', 'CUSAN.IS', 'DAGHL.IS', 'DAGI.IS', 'DAPGM.IS', 'DARDL.IS',  'DCTTR.IS', 'DEVA.IS', 'DGATE.IS', 'DENGE.IS', 'DERHL.IS', 'DERIM.IS', 'DESA.IS', 'DESPC.IS', 'DGNMO.IS', 'DGNMO.IS', 'DIRIT.IS', 'DITAS.IS', 'DMSAS.IS', 'DMRGD.IS', 'DNISI.IS', 'DOAS.IS', 'DOBUR.IS', 'DOCO.IS', 'DOGUB.IS', 'DOFER.IS', 'DOHOL.IS', 'DOKTA.IS', 'DURDO.IS', 'DURKN.IS', 'DZGYO.IS', 'ECILC.IS', 'ECZYT.IS', 'EDIP.IS', 'EFORC.IS', 'EGEEN.IS', 'EGGUB.IS', 'EGPRO.IS', 'EGSER.IS', 'EKGYO.IS', 'EKIZ.IS', 'ELITE.IS', 'EMNIS.IS', 'EMKEL.IS', 'ENJSA.IS', 'ENKAI.IS', 'ENSRI.IS', 'ERBOS.IS', 'EREGL.IS', 'ERSU.IS', 'ESCAR.IS', 'ESCOM.IS', 'ESEN.IS', 'ETILR.IS', 'ETYAT.IS', 'EUKYO.IS', 'EUREN.IS', 'EUYO.IS', 'FADE.IS', 'FENER.IS', 'FENER.IS','EDATA.IS',   'FLAP.IS', 'FMIZP.IS', 'FONET.IS', 'FORMT.IS', 'FORTE.IS', 'FRIGO.IS', 'FROTO.IS', 'FZLGY.IS', 'GARAN.IS', 'GARFA.IS', 'GEDIK.IS', 'GEDZA.IS', 'GENIL.IS', 'GENTS.IS', 'GESAN.IS', 'GIPTA.IS', 'GLBMD.IS', 'GLCVY.IS', 'GLRYH.IS', 'GLYHO.IS', 'GMTAS.IS', 'GOKNR.IS', 'GOLTS.IS', 'GOODY.IS', 'GOZDE.IS', 'GRNYO.IS', 'GRSEL.IS', 'GRTHO.IS', 'GSRAY.IS', 'GSDDE.IS', 'GSDHO.IS', 'GUBRF.IS', 'GUNDG.IS', 'GWIND.IS', 'GZNMI.IS', 'HALKB.IS',   'HATEK.IS', 'HATSN.IS', 'HDFGS.IS',  'HEDEF.IS', 'HEKTS.IS', 'HLGYO.IS',   'HKTM.IS', 'HOROZ.IS', 'HRKET.IS', 'HTTBT.IS', 'HUBVC.IS', 'HUNER.IS',  'HURGZ.IS',  'ICBCT.IS', 'ICUGS.IS', 'IDGYO.IS', 'IEYHO.IS', 'IHAAS.IS', 'IHEVA.IS', 'IHGZT.IS', 'IHLAS.IS', 'IHLGM.IS', 'IHYAY.IS', 'IMASM.IS', 'INDES.IS', 'INGRM.IS', 'INTEK.IS', 'INTEM.IS', 'INVEO.IS', 'INVES.IS',  'IPEKE.IS', 'ISATR.IS', 'ISBIR.IS', 'ISBTR.IS', 'ISCTR.IS', 'ISDMR.IS', 'ISFIN.IS', 'ISGYO.IS', 'ISGSY.IS', 'ISKPL.IS', 'ISKUR.IS', 'ISMEN.IS', 'ISSEN.IS', 'ISYAT.IS', 'IZENR.IS', 'IZFAS.IS', 'IZINV.IS', 'IZMDC.IS', 'JANTS.IS', 'KAPLM.IS', 'KAREL.IS', 'KARSN.IS', 'KARYE.IS', 'KARTN.IS', 'KATMR.IS', 'KAYSE.IS', 'KCAER.IS', 'KCHOL.IS', 'KFEIN.IS',  'KGYO.IS', 'KIMMR.IS', 'KLGYO.IS', 'KLKIM.IS', 'KLMSN.IS', 'KLRHO.IS', 'KLSER.IS',"KLNMA.IS", 'KLSYN.IS', 'KMPUR.IS', 'KNFRT.IS', 'KONKA.IS', 'KONTR.IS', 'KONYA.IS', 'KOPOL.IS', 'KORDS.IS', 'KRPLS.IS', 'KRONT.IS', 'KRSTL.IS', 'KRVGD.IS', 'KRTEK.IS', 'KRGYO.IS', 'KRDMA.IS', 'KRDMB.IS', 'KRDMD.IS', 'KSTUR.IS',  'KTLEV.IS', 'KUYAS.IS', 'KUVVA.IS', 'KZBGY.IS', 'KZGYO.IS', 'LIDFA.IS', 'LIDER.IS', 'LILAK.IS', 'LINK.IS', 'LKMNH.IS', 'LMKDC.IS', 'LOGO.IS', 'LRSHO.IS', 'LUKSK.IS', 'LYDHO.IS', 'MACKO.IS', 'MAGEN.IS', 'MAKIM.IS', 'MAKTK.IS', 'MAALT.IS', 'MANAS.IS', 'MARKA.IS', 'MARTI.IS', 'MAVI.IS', 'MEGAP.IS', 'MEGMT.IS', 'MEKAG.IS', 'MERCN.IS', 'MERIT.IS', 'MERKO.IS', 'METRO.IS', 'METUR.IS', 'MGROS.IS', 'MIATK.IS', 'MNDTR.IS', 'MNDRS.IS', 'MOBTL.IS', 'MOGAN.IS', 'MPARK.IS', 'MRGYO.IS', 'MRSHL.IS', 'MSGYO.IS', 'MTRKS.IS', 'MTRYO.IS', 'MZHLD.IS', 'NATEN.IS', 'NETAS.IS', 'NIBAS.IS', 'NTHOL.IS', 'NUHCM.IS', 'NUGYO.IS',  'ODAS.IS', 'ODINE.IS', 'OFSYM.IS', 'OBASE.IS', 'OBAMS.IS','ONCSM.IS', 'ONRYT.IS', 'ORCAY.IS', 'ORGE.IS', 'ORMA.IS', 'OSMEN.IS', 'OSTIM.IS', 'OTKAR.IS', 'OTTO.IS', 'OYAKC.IS', 'OYAYO.IS', 'OYLUM.IS', 'OZATD.IS', 'OZKGY.IS', 'OZGYO.IS', 'OZYSR.IS', 'OZRDN.IS', 'OZSUB.IS', 'PAGYO.IS', 'PAMEL.IS', 'PAPIL.IS', 'PARSN.IS', 'PASEU.IS', 'PATEK.IS', 'PCILT.IS', 'PEKGY.IS', 'PENGD.IS', 'PENTA.IS', 'PEHOL.IS', 'PETKM.IS', 'PETUN.IS', 'PGSUS.IS', 'PKART.IS', 'PKENT.IS', 'PKENT.IS', 'PKENT.IS', 'PKENT.IS', 'PNSUT.IS', 'POLHO.IS', 'POLTK.IS', 'PRDGS.IS', 'PRKME.IS', 'PRKAB.IS', 'PRZMA.IS', 'PSGYO.IS', 'PSDTC.IS', 'QNBFB.IS',  'QNBFL.IS', 'QUAGR.IS',   'RALYH.IS', 'RAYSG.IS', 'RNPOL.IS', 'RODRG.IS', 'ROYAL.IS', 'RTALB.IS', 'RUBNS.IS', 'RYGYO.IS', 'RYSAS.IS', 'SAFKR.IS', 'SAHOL.IS', 'SAMAT.IS', 'SANEL.IS', 'SANKO.IS', 'SANFM.IS', 'SARKY.IS', 'SASA.IS', 'SAYAS.IS', 'SELEC.IS', 'SEYKM.IS', 'SEKFK.IS', 'SEGYO.IS', 'SELEC.IS', 'SELGD.IS', 'SELVA.IS',  'SILVR.IS', 'SISE.IS', 'SKBNK.IS', 'SKTAS.IS', 'SMART.IS',   'SMRTG.IS', 'SNGYO.IS', 'SNICA.IS', 'SNKRN.IS', 'SODSN.IS', 'SOKE.IS', 'SONME.IS', 'SUMAS.IS', 'SUNTK.IS', 'SURGY.IS', 'SUWEN.IS', 'TABGD.IS',    'TAVHL.IS', 'TATGD.IS', 'TATEN.IS', 'TBORG.IS', 'TCELL.IS', 'TCKRC.IS',  'TDGYO.IS', 'TEKTU.IS', 'TERA.IS', 'THYAO.IS', 'TKFEN.IS','TEZOL.IS', 'TKNSA.IS', 'TLMAN.IS', 'TMSN.IS', 'TOASO.IS', 'TRCAS.IS', 'TRGYO.IS', 'TRILC.IS',      'TSKB.IS', 'TSGYO.IS', 'TSPOR.IS', 'TTKOM.IS', 'TTRAK.IS', 'TUPRS.IS', 'TURGG.IS', 'TUREX.IS', 'UFUK.IS', 'ULAS.IS', 'ULKER.IS', 'ULUFA.IS', 'ULUSE.IS', 'ULUUN.IS', 'UNLU.IS', 'USAK.IS', 'VAKBN.IS', 'VAKFN.IS', 'VAKKO.IS', 'VANGD.IS', 'VBTYZ.IS','VERUS.IS', 'VERTU.IS', 'VESBE.IS', 'VESTL.IS', 'VKFYO.IS', 'VKGYO.IS', 'VKING.IS', 'YAPRK.IS', 'YATAS.IS', 'YAYLA.IS', 'YAYLA.IS', 'YBTAS.IS', 'YEOTK.IS', 'YESIL.IS', 'YGYO.IS', 'YIGIT.IS', 'YKBNK.IS', 'YKSLN.IS', 'YONGA.IS', 'YUNSA.IS', 'YYAPI.IS', 'YYLGD.IS', 'ZEDUR.IS',  'ZOREN.IS', 'ZRGYO.IS']

# 2. Verileri çekmek için bir fonksiyon (en az 60 gün olacak şekilde)
def get_stock_data(ticker, period='3mo'):
    stock_data = yf.download(ticker, period=period)
    stock_data['Ticker'] = ticker
    stock_data.index = pd.to_datetime(stock_data.index, errors='coerce')
    return stock_data

# 3. Tüm hisselerin verilerini çek
data_list = [get_stock_data(ticker) for ticker in tickers]
all_stock_data = pd.concat(data_list)

# 4. RSI hesaplama fonksiyonu
def calculate_rsi(data, window=14):
    delta = data['Close'].diff()
    gain = delta.where(delta > 0, 0)
    loss = -delta.where(delta < 0, 0)
    avg_gain = gain.ewm(com=window - 1, min_periods=window).mean()
    avg_loss = loss.ewm(com=window - 1, min_periods=window).mean()
    rs = avg_gain / avg_loss
    rsi = 100 - (100 / (1 + rs))
    data["RSI"] = rsi
    return data

# 5. Bağıl hacmi hesaplama fonksiyonu
def calculate_relative_volume(data, period=10):
    data['Avg_Volume'] = data['Volume'].rolling(window=period).mean()
    data['Relative_Volume'] = data['Volume'] / data['Avg_Volume']
    return data

# 6. Fiyat değişimini hesaplama fonksiyonu
def calculate_price_change(data):
    data['Price_Change'] = data['Close'].pct_change() * 100
    return data

# 7. Fibonacci pivotlarını hesaplama fonksiyonu
def calculate_fibonacci_pivots(data):
    pivot_data = pd.DataFrame()
    data = data.reset_index()
    grouped = data.groupby('Ticker')
    for ticker, group in grouped:
        group = group.set_index('Date')
        monthly_data = group.resample('M').agg({'High': 'max', 'Low': 'min', 'Close': 'last'}) # Bir önceki ayın verilerini kullanmak için
        monthly_data = monthly_data.dropna()
        if monthly_data.empty:
            continue  # Eğer yeterli veri yoksa atla
        high = monthly_data['High']
        low = monthly_data['Low']
        close = monthly_data['Close']
        pp = (high + low + close) / 3
        r1 = pp + (high - low) * 0.382
        r2 = pp + (high - low) * 0.618
        r3 = pp + (high - low) * 1.0
        s1 = pp - (high - low) * 0.382
        s2 = pp - (high - low) * 0.618
        s3 = pp - (high - low) * 1.0
        pivot_df = pd.DataFrame({
            'Date': monthly_data.index + pd.offsets.MonthBegin(1),  # Tarihi ayın ilk gününe ayarlıyoruz
            'Ticker': ticker,
            'Pivot': pp.values,
            'R1': r1.values,
            'R2': r2.values,
            'R3': r3.values,
            'S1': s1.values,
            'S2': s2.values,
            'S3': s3.values
        })
        pivot_df['Month'] = pivot_df['Date'].dt.to_period('M')  # Ay sütunu ekliyoruz
        pivot_data = pd.concat([pivot_data, pivot_df], ignore_index=True)
    return pivot_data

# 8. Tüm hesaplamaları uygulayan fonksiyon
def apply_calculations(data):
    data = data.groupby('Ticker', group_keys=False).apply(calculate_rsi)
    data = data.groupby('Ticker', group_keys=False).apply(calculate_relative_volume)
    data = data.groupby('Ticker', group_keys=False).apply(calculate_price_change)
    return data

# 9. Pivot verilerini günlük verilere ekleyen fonksiyon
def merge_pivots_with_daily_data(all_stock_data, pivot_data):
    # Tarihleri datetime formatına çeviriyoruz
    all_stock_data['Date'] = pd.to_datetime(all_stock_data.index)
    pivot_data['Date'] = pd.to_datetime(pivot_data['Date'])
    # Ay sütunlarını ekliyoruz
    all_stock_data['Month'] = all_stock_data['Date'].dt.to_period('M')
    pivot_data['Month'] = pivot_data['Date'].dt.to_period('M')
    # İndeksleri sıfırlıyoruz
    all_stock_data.reset_index(drop=True, inplace=True)
    pivot_data.reset_index(drop=True, inplace=True)
    # Verileri 'Ticker' ve 'Month' sütunlarına göre birleştiriyoruz
    all_stock_data = pd.merge(all_stock_data, pivot_data[['Ticker', 'Month', 'Pivot', 'R1', 'R2', 'R3', 'S1', 'S2', 'S3']], on=['Ticker', 'Month'], how='left')
    # Verileri sıralıyoruz
    all_stock_data.sort_values(['Ticker', 'Date'], inplace=True)
    return all_stock_data

# 10. Hisseleri filtreleyen fonksiyon
def filter_stocks(data):
    filtered_stocks = []
    for ticker in data['Ticker'].unique():
        stock = data[data['Ticker'] == ticker].tail(20)
        if len(stock) < 2:
            continue  # Yeterli veri yoksa atla
        last_day = stock.iloc[-2]
        previous_day = stock.iloc[-3]
        # RSI kriteri (50-60 arası)
        rsi_criteria = 50 <= last_day['RSI'] <= 60
        # Bağıl Hacim kriteri (1.5 ve üzeri)
        relative_volume_criteria = last_day['Relative_Volume'] >= 1.5
        # Pivot geçiş kriteri
        pivot_crossing_criteria = (previous_day['Close'] < previous_day['Pivot']) and (last_day['Close'] > last_day['Pivot'])
        # Fiyat değişimi kriteri (%4.99 ve üzeri)
        #price_change_criteria = last_day['Price_Change'] >= 4.99
        # Tüm kriterler sağlanıyorsa hisseyi ekle
        if rsi_criteria and relative_volume_criteria and  pivot_crossing_criteria:
            filtered_stocks.append(last_day)
    return pd.DataFrame(filtered_stocks)

# 11. Tüm verilerin işlenmesi
all_stock_data = apply_calculations(all_stock_data)

# 12. Fibonacci pivotlarını hesaplıyoruz
monthly_pivots = calculate_fibonacci_pivots(all_stock_data)

# 13. Pivot verilerini günlük verilere ekliyoruz
all_stock_data = merge_pivots_with_daily_data(all_stock_data, monthly_pivots)

# 14. Filtreleme işlemi
filtered_stocks = filter_stocks(all_stock_data)

# 15. Filtrelenen sonuçları gösterme
if not filtered_stocks.empty:
    print(filtered_stocks[['Ticker', 'Date', 'Close', 'RSI', 'Relative_Volume', 'Pivot', 'S1', 'Price_Change']])
else:
    print("Filtrelenen sonuçlar boş. Kriterlere uyan hisse bulunamadı.")
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import matplotlib.pyplot as plt
import math
from sklearn.metrics import mean_squared_error

# Filtrelenen hisselerin listesi (önceki kodunuzdan elde edilen)
filtered_tickers = filtered_stocks['Ticker'].unique()
print("Filtrelenen hisseler:", filtered_tickers)

def get_lstm_data(ticker, period='1y'):
    stock_data = yf.download(ticker, period=period)
    stock_data = stock_data[['Close']]
    stock_data.dropna(inplace=True)
    return stock_data

def create_dataset(dataset, time_step=60):
    X, Y = [], []
    for i in range(len(dataset) - time_step - 1):
        X.append(dataset[i:(i + time_step), 0])
        Y.append(dataset[i + time_step, 0])
    return np.array(X), np.array(Y)

# Her hisse için modeli oluştur ve eğit
for ticker in filtered_tickers:
    print(f"\n{ticker} için LSTM modeli oluşturuluyor...")
    data = get_lstm_data(ticker)
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(data)
    time_step = 60
    X, Y = create_dataset(scaled_data, time_step)
    train_size = int(len(X) * 0.8)
    test_size = len(X) - train_size
    X_train, X_test = X[0:train_size], X[train_size:len(X)]
    Y_train, Y_test = Y[0:train_size], Y[train_size:len(Y)]
    X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
    X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)
    model = Sequential()
    model.add(LSTM(50, return_sequences=True, input_shape=(time_step, 1)))
    model.add(LSTM(50))
    model.add(Dense(1))
    model.compile(loss='mean_squared_error', optimizer='adam')
    model.fit(X_train, Y_train, validation_data=(X_test, Y_test), epochs=50, batch_size=64, verbose=1)
    train_predict = model.predict(X_train)
    test_predict = model.predict(X_test)
    train_predict = scaler.inverse_transform(train_predict.reshape(-1, 1))
    Y_train_actual = scaler.inverse_transform(Y_train.reshape(-1, 1))
    test_predict = scaler.inverse_transform(test_predict.reshape(-1, 1))
    Y_test_actual = scaler.inverse_transform(Y_test.reshape(-1, 1))
    train_rmse = math.sqrt(mean_squared_error(Y_train_actual, train_predict))
    test_rmse = math.sqrt(mean_squared_error(Y_test_actual, test_predict))
    print(f"{ticker} - Eğitim Hatası (RMSE): {train_rmse}")
    print(f"{ticker} - Test Hatası (RMSE): {test_rmse}")
    # Sonuçları görselleştirebilirsiniz
    plt.figure(figsize=(12, 6))
    plt.plot(Y_test_actual, label='Gerçek Değerler')
    plt.plot(test_predict, label='Tahminler')
    plt.title(f"{ticker} - Test Sonuçları")
    plt.legend()
    plt.show()
```
